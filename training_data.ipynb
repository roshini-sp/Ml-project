```python
data = []
labels = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".wav"):
            try:
                file_path = os.path.join(root, file)

                feature = extract_features(file_path)
                data.append(feature)

                # RAVDESS
                if "Actor" in root:
                    label = file.split("-")[2]

                # TESS
                else:
                    label = file.split("_")[-1].replace(".wav", "")

                labels.append(label)

            except Exception as e:
                print("Error:", file_path, e)

print("Total samples:", len(data))
```

```python
emotion_map = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fear",
    "07": "disgust",
    "08": "surprise"
}

labels = [emotion_map[label] if label in emotion_map else label for label in labels]

print("Unique labels:", set(labels))
```

```python
X = np.array(data)
y = np.array(labels)
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

le = LabelEncoder()
y_encoded = to_categorical(le.fit_transform(y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

model = Sequential()

model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.3))

model.add(LSTM(64))
model.add(Dropout(0.3))

model.add(Dense(32, activation='relu'))
model.add(Dense(y_encoded.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
def extract_features(file_path):
    audio, sample_rate = librosa.load(file_path, duration=3, offset=0.5)

    mfcc = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40).T, axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(y=audio, sr=sample_rate).T, axis=0)
    mel = np.mean(librosa.feature.melspectrogram(y=audio, sr=sample_rate).T, axis=0)
    zcr = np.mean(librosa.feature.zero_crossing_rate(audio).T, axis=0)

    return np.hstack([mfcc, chroma, mel, zcr])
```

```python
data = []
labels = []

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".wav"):
            try:
                file_path = os.path.join(root, file)

                feature = extract_features(file_path)
                data.append(feature)

                if "Actor" in root:
                    label = file.split("-")[2]
                else:
                    label = file.split("_")[-1].replace(".wav", "")

                labels.append(label)

            except Exception as e:
                print("Error:", file_path, e)

print("Total samples:", len(data))
```

```python
emotion_map = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fear",
    "07": "disgust",
    "08": "surprise"
}

labels = [emotion_map[label] if label in emotion_map else label for label in labels]

print("Unique labels:", set(labels))
```

```python
from sklearn.preprocessing import StandardScaler

X = np.array(data)
y = np.array(labels)

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
le = LabelEncoder()
y_encoded = to_categorical(le.fit_transform(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
model = Sequential()

model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.4))

model.add(LSTM(64))
model.add(Dropout(0.4))

model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(y_encoded.shape[1], activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss', patience=5)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[early_stop]
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print("Final Accuracy:", accuracy)
```
